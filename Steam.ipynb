# Merging 2 DataFrames ( CSV FILES )

```python
import pandas as pd
```

```python
import pandas as pd

var1 = pd.DataFrame({"Name": ["Shub", "Jimmy", "Joe", "Travis", "Paul"], "Salary": [1050, 1200, 1293, 5052, 90424]})
var1
```

```python
var2 = pd.DataFrame({"Name": ["Shub", "Jimmy", "Joe", "Travis", "Paul"],
                     "Role": ["Administrative Assistant", "Office Manager", "HR Manager", "IT Specialist", "CEO"],
                     "Contact": ["shub@icloud.com", "Jimmy@gmail.com", "joe@microsoft.com", "Travis@linux.com", "Paul@icloud.com"]})  
var2
```

```python
merged = pd.merge(var1, var2, on="Name",indicator = True)       # Merge DataFrames with indicator value

merged
```
